```python
def categorize_performance(sales_amount):
    if sales_amount > 10000:
        return "Excellent"
    elif 5000 <= sales_amount <= 10000:
        return "Good"
    elif 2000 <= sales_amount < 5000:
        return "Average"
    else:
        return "Poor"

def main():
    sales_amount = float(input("Enter the sales amount: "))
    performance = categorize_performance(sales_amount)
    print("Employee's performance:", performance)

if __name__ == "__main__":
    main()
```

```python
sentence = "Torture the data, and it will confess to anything"
# 1. Count the length of the string
length = len(sentence)
print("1. Length of the string:", length)
# 2. Reverse the string
reversed_sentence = sentence[::-1]
print("2. Reversed string:", reversed_sentence)
# 3. Replace ‘anything’ with ‘everything’ and print both the sentences
new_sentence = sentence.replace("anything", "everything")
print("3. Original sentence:", sentence)
print("   Replaced sentence:", new_sentence)
# 4. Extract 'torture' and 'confess'
words = sentence.split()
torture = words[0]
confess = words[-3]
print("4. Extracted words: Torture =", torture + ",", "Confess =", confess)
# 5. Check if the sentence starts with a vowel or consonant
vowels = ['a', 'e', 'i', 'o', 'u']
first_char = sentence[0].lower()
if first_char in vowels:
    print("5. The sentence starts with a vowel.")
else:
    print("5. The sentence starts with a consonant.")
```

```python
def divide_with_remainder(a, b):
    quotient = a // b  
    remainder = a % b  
    return (quotient, remainder)

# Example usage:
a = 23
b = 5
result = divide_with_remainder(a, b)
print("Quotient:", result[0])
print("Remainder:", result[1])
```

```python
def calculate_operations(y, m, o, z):
    addition_result = y + m + o + z
    subtraction_result = y - m - o - z
    square_results = {
        'y^2': y ** 2,
        'm^2': m ** 2,
        'o^2': o ** 2,
        'z^2': z ** 2
    }
    if z != 0:
        division_result = y / z
    else:
        division_result = "Division by zero is undefined."
    
    return addition_result, subtraction_result, square_results, division_result
y = int(input("Enter the value of y: "))
m = int(input("Enter the value of m: "))
o = int(input("Enter the value of o: "))
z = int(input("Enter the value of z: "))
add_result, sub_result, square_result, div_result = calculate_operations(y, m, o, z)
print("Addition Result:", add_result)
print("Subtraction Result:", sub_result)
print("Square Results:", square_result)
print("Division Result:", div_result)
```

```python
import matplotlib.pyplot as plt 
import pandas as pd
import seaborn as sns
```

```python
sales_data['Profit_Margin'] = (sales_data['Profit'] / sales_data['Revenue']) * 100

plt.figure(figsize=(10, 6))
plt.hist(sales_data['Profit_Margin'], bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Profit Margin')
plt.xlabel('Profit Margin (%)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Customer_Age', y='Profit', data=sales_data)
plt.title('Relationship between Customer Age and Profit')
plt.xlabel('Customer Age')
plt.ylabel('Profit')
plt.grid(True)
plt.show()
```

```python
if 'Year' in sales_data.columns and 'Month' in sales_data.columns:
    order_quantity_over_time = sales_data.groupby(['Year', 'Month'])['Order_Quantity'].sum().reset_index()
    plt.figure(figsize=(12, 6))
    sns.lineplot(x='Month', y='Order_Quantity', hue='Year', data=order_quantity_over_time, marker='o')
    plt.title('Trend of Order Quantity Over Time')
    plt.xlabel('Month')
    plt.ylabel('Total Order Quantity')
    plt.grid(True)
    plt.show()
else:
    print("Time data is not available.")
```
